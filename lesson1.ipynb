```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A partir de uma lista
array = np.array([1,2,3,4,5])
print("Meu array:", array)

# Criando um array de zeros
zeros = np.zeros((2, 3)) # Matriz de 2x3 de zeros
print("Array de zeros:\n", zeros)

# Criando um array com um intervalo
interval = np.arange(0,10,2) # de 0 a 10, de 2 em 2
print("Array com intervalo:\n", interval)
```

```python
array_a = np.array([1,2,3])
array_b = np.array([4,5,6])

sum = array_a + array_b
multiplication = array_a * 2

print("Soma dos arrays:\n", sum)
print("Multiplicação array a:\n", multiplication)
```

```python
years = [2020,2021,2022,2023,2024]
sells = [100,120,150,130,160]

plt.plot(years,sells)
plt.title("Vendas ao Longo dos Anos")
plt.xlabel("Ano")
plt.ylabel("Vendas (em milhões)")
plt.show()
```

```python
products = ["Produto A", "Produto B", "Produto C"]
quantities = [50,80,65]

plt.bar(products, quantities)
plt.title("Quantidade de Produtos em Estoque")
plt.xlabel("Produto")
plt.ylabel("Quantidade")
plt.show()
```

```python
# Gerando dados com Numpy
x = np.linspace(0,2 * np.pi, 100)
y_seno = np.sin(x)
y_cosseno = np.cos(x)

# Criando o gráfico com matplotlib
plt.plot(x, y_seno, label="Seno")
plt.plot(x, y_cosseno, label="Cosseno")
plt.title("Funções Seno e Cosseno")
plt.xlabel("Ângulo [rad]")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.show()
```
